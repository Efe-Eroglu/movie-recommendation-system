```python
import pandas as pd
import numpy as np
```

```python
name = pd.read_csv("/content/name.basics.tsv",sep="\t",low_memory=False)
title = pd.read_csv("/content/title.basics.tsv",sep="\t",low_memory=False)
title_crew = pd.read_csv("/content/title.crew.tsv",sep="\t",low_memory=False)
title_ratings = pd.read_csv("/content/title.ratings.tsv",sep="\t",low_memory=False)
```

```python
name.dropna(axis="index", inplace=True)
title.dropna(axis="index", inplace=True)
title_crew.dropna(axis="index", inplace=True)
title_ratings.dropna(axis="index", inplace=True)
```

```python
title = title[['tconst', 'primaryTitle', 'originalTitle', 'startYear', 'genres']]
title_ratings = title_ratings[['tconst', 'averageRating', 'numVotes']]
title_crew = title_crew[['tconst', 'directors']]
name = name[['nconst', 'primaryName']]
```

```python
title.head()
```

```python
title = title.rename(columns={
    "tconst" : "movie_id",
    "primaryTitle":"movie_title",
    "originalTitle":"original_title",
    "startYear":"release_year",
    "genres":"movie_genres"
})
```

```python
title_ratings.head()
```

```python
title_ratings = title_ratings.rename(columns = {
    "tconst" : "movie_id",
    "averageRating" :"rating",
    "numVotes": "votes"
})
```

```python
title_crew.head()
```

```python
title_crew = title_crew.rename(columns={
    "tconst" : "movie_id",
})
```

```python
name.head()
```

```python
name = name.rename(columns={
  "nconst" : "person_id",
  "primaryName":"person_name"
})
```

```python
title = title.replace(r'\\N', pd.NA, regex=True)
title_ratings = title_ratings.replace(r'\\N', pd.NA, regex=True)
title_crew = title_crew.replace(r'\\N', pd.NA, regex=True)
name = name.replace(r'\\N', pd.NA, regex=True)
```

```python
merged_df = pd.merge(title, title_ratings, on="movie_id", how = "inner")
```

```python
merged_df = pd.merge(merged_df, title_crew, on="movie_id", how="left")
```

```python
merged_df
```

```python
title_crew['directors'] = title_crew['directors'].str.split(',')
```

```python
title_crew
```

```python
title_crew_exploded = title_crew.explode('directors')

title_crew_exploded = title_crew_exploded.merge(name, left_on='directors', right_on='person_id', how='left')

directors_grouped = title_crew_exploded.groupby('movie_id')['person_name'].apply(list).reset_index()
directors_grouped = directors_grouped.rename(columns={'person_name': 'director_names'})

merged_df = pd.merge(merged_df, directors_grouped, on='movie_id', how='left')
```

```python
df = merged_df
```

```python
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

df['movie_genres'] = df['movie_genres'].str.split(',')
```

```python
df = df.dropna()
```
